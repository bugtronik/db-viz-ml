```python
# Beta tests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import de la data source et affichage des donn√©es de tests
df_sales_summary = pd.read_csv('../datasources/all_sales_by_products.csv')
print(df_sales_summary.head())
```

```python
# Affichage du graphs de la synth√®se des ventes mobipass sur l'ensemble de l'ann√©e 2020
sns.barplot(
    data=df_sales_summary,
    x='PRESTATION',
    y='CA HT',
    hue="PRESTATION"
)
plt.title("Chiffre d'affaire par prestation en 2020")
plt.xlabel('Prestations')
plt.ylabel('CA HT(Francs CFA)')

plt.show()
```
